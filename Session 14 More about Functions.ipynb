## 3

    1. Write a function a takes a string
    2. Check if the string contains a vowel ( aeiou )
    3. If it does, print the string once 
    4. else print the string n times (n should be passed by the user)
    
Intuition :
    1. Write a function to check if the word contains vowel
    2. Use the output from above to print desired number of times

```python
# 7:55 (solve 3 and feedback)
```

```python
def checkifvowels(string):
    vowels = "aeiou"
    for ch in string:
        if ch in vowels:
            return True
    else:
        return False
```

```python
checkifvowels("mht")
```

```python
def print_string_n_times(string,n):
    for i in range(n):
        print(string)
```

```python
def task(word,h):
    check = checkifvowels(word)
    if check==True:
        print(word)
    else:
        print_string_n_times(word,h)
```

```python
n = int(input())
string = input()

task(string,n)
```

## Variable Arguments

* Packing

```python
# we are just using args as name, we can use any name here. But everybody use args.

def mysum(*args):
    print(type(args),args)
    res = 0
    for value in args:
        res = res+value
    return res
```

```python
mysum(1,2,3,4,5,6,7,8)
```

## Named Packing

```python
* =>tuple, **=>dict
```

```python
# we have to pass arguments in order
# we are just using kwargs as name, we can use any name here. But everybody use kwargs.


def task(**kwargs):
    print(type(kwargs),kwargs)
    print(kwargs['name'],"is studying at ",kwargs['college'],"he took this stream",kwargs['stream'])
```

```python
task(college="MIT",name="rohan rathore",stream="Computer Science")
```
