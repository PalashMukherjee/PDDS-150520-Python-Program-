```python
tell() and seek() in file operations

tell(): finds the current fp position
seek(): change the position oof the fp to the desired postion
```

```python
# SEEK OP. WITH RESPECT TO FILE OPENING IN R+,W, IN THSE MODES, NOT IN BINARY MODES LIKE rb+
# fp.seek(offset,position)

# postion are: 
        # os.SEEK_SET  (Beginning OF THE FILE, OFFSET CAN BE +VALUES HERE)
        # os.SEEK_END  (Ending OF THE FILE, limited use, OFFSET CAN BE 0 HERE)
```

```python
# Normal Seek Operations

import os

fp = open("test1.txt","r+")
print("1file pointer is at position",fp.tell())

content = list(fp.readline())
print(content)
print("2file pointer is at position",fp.tell())

fp.seek(1,os.SEEK_SET)
print("3file pointer is at position",fp.tell())

fp.seek(10,os.SEEK_SET)
print("4file pointer is at position",fp.tell())

fp.seek(0,os.SEEK_END)
print("5file pointer is at position",fp.tell())

fp.seek(20,os.SEEK_SET)
print("6file pointer is at position",fp.tell())
```

```python
# only relative seeks,offsets allowed with os.SEEK_SET
fp.seek(10,os.SEEK_END) # error ecause we used os.SEEK_END
print("5file pointer is at position",fp.tell())
```

```python
# Smart Seek Operations

# only fp.tell() used in offsets are legal

import os

fp = open("test1.txt","r+")
print("1file pointer is at position",fp.tell())

content = list(fp.readline())
print(content)
print("2file pointer is at position",fp.tell())

# fp.seek(1,os.SEEK_SET)
fp.seek(fp.tell()-5,os.SEEK_SET)
print("3file pointer is at position",fp.tell())

# fp.seek(10,os.SEEK_SET)
fp.seek(fp.tell()+9,os.SEEK_SET)
print("4file pointer is at position",fp.tell())

fp.seek(0,os.SEEK_END)
print("5file pointer is at position",fp.tell())

fp.seek(fp.tell()-3,os.SEEK_SET)
print("6file pointer is at position",fp.tell())
```

```python
## backreferencing

regex = r'([a-z]+),\1'

print(re.search(regex,"cat,cat,cat"))

print(re.findall(regex,"cat,cat,car,car"))
```

```python
# infinite sequence of natural numbers

def natural_no():
    n=1
    while True:
        yield n
        n=n+1  # n+=1
        
iterator_obj = natural_no()
```

```python
print(next(iterator_obj))
```

```python
for i in range(10):
    print(next(iterator_obj))
```
